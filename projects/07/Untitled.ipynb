```python
from pprint import pprint


memory = {
    'stack': [],
    'local': [0]*16,
    'argument': [0]*16,
    'this': None,
    'that': None,
    'temp': [0]*8
}
memory
```

```python
#
# ARITHMETIC
#
    
def eq():
    y = stack.pop()
    x = stack.pop()
    stack.append(-1 if x==y else 0)
    pprint(stack)
    print('> EQ DONE')

def lt():
    y = stack.pop()
    x = stack.pop()
    stack.append(-1 if x<y else 0)
    pprint(stack)
    print('> LT DONE')

def gt():
    y = stack.pop()
    x = stack.pop()
    stack.append(-1 if x>y else 0)
    pprint(stack)
    print('> GT DONE')
    
def add():
    y = memory['stack'].pop()
    x = memory['stack'].pop()
    memory['stack'].append(x+y)
#     pprint(stack)
    print('> ADD DONE')
    
def sub():
    y = memory['stack'].pop()
    x = memory['stack'].pop()
    memory['stack'].append(x-y)
#     pprint(stack)
    print('> SUB DONE')
    
def neg():
    y = stack.pop()
    stack.append(-y)
    pprint(stack)
    print('> NEG DONE')
    
def _and():
    y = stack.pop()
    x = stack.pop()
    stack.append(x&y)
    pprint(stack)
    print('> AND DONE')
    
def _or():
    y = stack.pop()
    x = stack.pop()
    stack.append(x|y)
    pprint(stack)
    print('> OR DONE')
    
def _not():
    y = stack.pop()
    stack.append(~y)
    pprint(stack)
    print('> NOT DONE')
```

```python
#
# MEMORY
#
    
def _push(segment, index):
    # Segments
    if segment == 'constant':
        x = index
    elif segment == 'local':
        x = memory['local'][index]
        memory['local'][index] = 0
    elif segment == 'argument':
        x = memory['argument'][index]
        memory['argument'][index] = 0
    elif segment == 'temp':
        x = memory['temp'][index]
        memory['temp'][index] = 0 
    elif segment == 'this':
        x = memory['this']
        memory['this'] = 0
    elif segment == 'that':
        x = memory['that']
        memory['that'] = 0
    else:
        print("################# BROKEN PUSH {} ############".format(segment))

    # Stack
    print('\t\t> PUSH {} {}'.format(segment, index))
    memory['stack'].append(x)

def _pop(segment, index):
    # Stack
    print('\t\t> POP {} {}'.format(segment, index))
    x = memory['stack'].pop()

    # Segments
    if segment == 'local':
        memory['local'][index] = x
    elif segment == 'argument':
        memory['argument'][index] = x
    elif segment == 'temp':
        memory['temp'][index] = x
    elif segment == 'this':
        memory['this'] = x
    elif segment == 'that':
        memory['that'] = x
    else:
        print("################# BROKEN POP {} ############".format(segment))
    
```

```python


program = """
push constant 10    // push 17 onto stack
pop local 0         // pop into RAM[local + 0]
push constant 21    // push 21 onto stack
push constant 22    // push 22 onto stack
pop argument 2      // pop into RAM[argument + 2]
pop argument 1      // pop into RAM[argument + 1]
push constant 36    // push 36 onto stack
pop this 6          // pop into RAM[this + 6]
push constant 42    // push 42 onto stack
push constant 45    // push 45 onto stack
pop that 5          // pop into RAM[that + 5]
pop that 2          // pop into RAM[that + 2]
push constant 510   // push 510 onto stack
pop temp 6          // pop stack to RAM[5 (temp) + 6]
push local 0        // pop into RAM[local + 0]
push that 5         // push from RAM[that + 5]
add                 // Command
push argument 1     // push from RAM[argument + 1]
sub                 // Command
push this 6         // push from RAM[this + 6]
push this 6         // push from RAM[this + 6]
add                 // Command
sub                 // Command
push temp 6         // push from RAM[5 (temp) + 6]
add                 // Command
"""
program = [x for x in program.split('\n') if x != '']


memory = {
    'stack': [],
    'local': [0]*16,
    'argument': [0]*16,
    'this': 0,
    'that': 0,
    'temp': [0]*8
}

for line in program:
    args = line.split()
    if args[0] == 'push':
        _push(args[1], int(args[2]))
    elif args[0] == 'pop':
        _pop(args[1], int(args[2]))
    elif args[0] == 'add':
        add()
    elif args[0] == 'sub':
        sub()
    else:
        print("***** {} *****".format(args[0]))
        
    pprint(memory)
    print()
```
