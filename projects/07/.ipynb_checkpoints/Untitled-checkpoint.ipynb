```python
#
# ARITHMETIC
#
    
def eq():
    y = stack.pop()
    x = stack.pop()
    stack.append(-1 if x==y else 0)
    pprint(stack)
    print('> EQ DONE')

def lt():
    y = stack.pop()
    x = stack.pop()
    stack.append(-1 if x<y else 0)
    pprint(stack)
    print('> LT DONE')

def gt():
    y = stack.pop()
    x = stack.pop()
    stack.append(-1 if x>y else 0)
    pprint(stack)
    print('> GT DONE')
    
def add():
    y = stack.pop()
    x = stack.pop()
    stack.append(x+y)
    pprint(stack)
    print('> ADD DONE')
    
def sub():
    y = stack.pop()
    x = stack.pop()
    stack.append(x-y)
    pprint(stack)
    print('> SUB DONE')
    
def neg():
    y = stack.pop()
    stack.append(-y)
    pprint(stack)
    print('> NEG DONE')
    
def _and():
    y = stack.pop()
    x = stack.pop()
    stack.append(x&y)
    pprint(stack)
    print('> AND DONE')
    
def _or():
    y = stack.pop()
    x = stack.pop()
    stack.append(x|y)
    pprint(stack)
    print('> OR DONE')
    
def _not():
    y = stack.pop()
    stack.append(~y)
    pprint(stack)
    print('> NOT DONE')
```
